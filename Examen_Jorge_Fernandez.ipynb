## Pregunta 1 : Explica como el algoritmo K-Nearest Neighbors utiliza metricas de similitud para generar recomendaciones. Indica dos ventajas y dos limitaciones del uso de KNN en un sistema de recomendacion con datos de peliculas.

###  Respuesta

KNN en sistemas de recomendación identifica elementos similares a un usuario o un ítem (película) en función de métricas de similitud, como la distancia euclidiana, la similitud de coseno, o la correlación de Pearson. En un sistema de recomendación de películas, el algoritmo puede buscar usuarios con perfiles similares o películas con características o ratings parecidos. Basándose en estas similitudes, KNN recomienda películas populares entre los usuarios más similares o películas con un perfil similar a las que un usuario ya ha disfrutado.

** Ventajas de KNN:

- Fácil de interpretar: Es intuitivo y fácil de implementar, pues recomienda películas en función de la cercanía a los ítems o usuarios más similares.
- No requiere entrenamiento pesado: Al ser un modelo basado en instancias, no necesita pre-entrenamiento intensivo.

** Limitaciones de KNN:

- Escalabilidad: Puede volverse computacionalmente costoso en datasets grandes, ya que debe calcular la distancia entre muchos pares de usuarios o películas.
- Escasez de datos: La precisión puede ser baja si hay poca información (ratings) compartida entre usuarios, lo que afecta la calidad de las recomendaciones.

Pregunta 2 : Selecciona dos usuarios al azar que hayan valorado al menos 5 peliculas en comun. Calcula la similitud entre estos dos usuarios usando la correlacion de Pearson. Luego, interpreta el resultado y discute que indicaria esta similitud para un sistema que busca recomendar peliculas.

```python
from itertools import combinations
df = pd.DataFrame(user_item_rating)

# Paso 1: Crear una tabla pivote con los usuarios y las películas que calificaron
pivot = df.pivot_table(index='userId', columns='title', values='rating', aggfunc='first')

# Paso 2: Encontrar pares de usuarios que tengan al menos 5 películas en común
user_pairs = []
for user1, user2 in combinations(pivot.index, 2):
    # Contar cuántas películas ambos usuarios han calificado
    common_movies = pivot.loc[user1].notna() & pivot.loc[user2].notna()
    if common_movies.sum() >= 5:
        user_pairs.append((user1, user2))
```

Pregunta 4: Implementa un sistema de recomendacion basado en KNN utilizando la distancia Euclideanda como metrica de similitud. Divide el datased en conjuntos de entrenamineto y prueba.Luego , calcula el hit rate del modelo en el conjunto de prueba, explicando como el hit rate refleja la efectividad del recomendador en este caso.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import pairwise_distances
import numpy as np

pivot = user_item_rating.pivot_table(index='userId', columns='title', values='rating', fill_value=0)
pivot
```

```python
def Random_Holdout(df, test_size=0.25):
    """Divide los datos asegurando que cada usuario tenga al menos una interacción en el conjunto de entrenamiento"""
    # Seleccionar al azar un porcentaje de interacciones por usuario para entrenamiento
    train_df = df.groupby('userId').apply(lambda x: x.sample(frac=1 - test_size, random_state=42)).reset_index(drop=True)
    # El resto va al conjunto de prueba
    test_df = pd.concat([df, train_df]).drop_duplicates(keep=False)

    return train_df, test_df

# Aplicar la función al dataset filtrado
train_df, test_df = Random_Holdout(user_item_rating, test_size=0.3)
```

```python
from surprise import Dataset, Reader, KNNBasic, accuracy

# Definir el rango de calificaciones
reader = Reader(rating_scale=(1, 5))

# Cargar datos de entrenamiento
trainset = Dataset.load_from_df(train_df[['userId', 'title', 'rating']], reader).build_full_trainset()

# Crear conjunto de prueba
testset = list(test_df[['userId', 'title', 'rating']].itertuples(index=False, name=None))
```

```python

# Configurar opciones del modelo KNN
sim_options = {
    'name': 'msd',  # Mean Squared Difference (distancia euclidiana)
    'user_based': True  # Filtrado Usuario-Usuario
}

# Crear el modelo KNN
knn = KNNBasic(k=50, sim_options=sim_options)

# Entrenar el modelo
knn.fit(trainset)
```

```python

# Hacer predicciones sobre el conjunto de prueba
predictions = knn.test(testset)

# Calcular el RMSE
rmse = accuracy.rmse(predictions)
print(f"RMSE del modelo: {rmse:.4f}")
```

```python
predictions_df = pd.DataFrame(predictions, columns=['userId', 'movieId', 'real_rating', 'predicted_rating', 'details'])
```

## Hit Rate Global

```python
# Definir umbral de relevancia
relevant_threshold = 4.0

# Inicializar contadores
hits = 0
total_relevant = 0

# Iterar sobre todas las predicciones
for _, row in predictions_df.iterrows():
    real_rating = row['real_rating']
    predicted_rating = row['predicted_rating']

    # Considerar películas con calificación real ≥ umbral como relevantes
    if real_rating >= relevant_threshold:
        total_relevant += 1
        # Si la predicción también es ≥ umbral, es un acierto
        if predicted_rating >= relevant_threshold:
            hits += 1

# Calcular el Hit Rate Global
hit_rate_global = hits / total_relevant if total_relevant > 0 else 0
print(f"Hit Rate Global: {hit_rate_global:.2f}")
```

## Hit Rate por usuario

```python

# Inicializar lista para almacenar hit rates por usuario
hit_rates = []

# Agrupar predicciones por usuario
for user_id, group in predictions_df.groupby('userId'):
    hits = 0
    total_relevant = 0

    for _, row in group.iterrows():
        real_rating = row['real_rating']
        predicted_rating = row['predicted_rating']

        # Considerar películas relevantes
        if real_rating >= relevant_threshold:
            total_relevant += 1
            if predicted_rating >= relevant_threshold:
                hits += 1

    # Calcular Hit Rate para el usuario
    if total_relevant > 0:
        hit_rate_user = hits / total_relevant
        hit_rates.append(hit_rate_user)

# Calcular el Hit Rate promedio
hit_rate_avg_user = sum(hit_rates) / len(hit_rates) if len(hit_rates) > 0 else 0
print(f"Hit Rate promedio por usuario: {hit_rate_avg_user:.2f}")
```

- Eficiencia del Modelo: Un Hit Rate global del 46% sugiere que el modelo tiene una capacidad razonable para presentar recomendaciones que los usuarios encuentran útiles. Esto implica que el sistema puede identificar patrones en las preferencias de los usuarios y, por ende, hacer recomendaciones que son relevantes para una parte significativa de ellos.

- Relevancia Promedio: El Hit Rate promedio por usuario de 48% sugiere que, a nivel individual, los usuarios tienden a interactuar positivamente con casi la mitad de las recomendaciones. Esto refuerza la idea de que el modelo tiene una buena comprensión de las preferencias del usuario, aunque hay espacio para mejorar.

- Áreas de Mejora: A pesar de los resultados positivos, un Hit Rate de 46% indica que hay un 54% de las recomendaciones que no son bien recibidas. Esto representa una oportunidad para ajustar el modelo, mejorar la calidad de las recomendaciones y optimizar el sistema para que sea más preciso. Se podrían explorar técnicas adicionales, como la incorporación de más datos o la personalización de las recomendaciones según los intereses específicos de cada usuario.
